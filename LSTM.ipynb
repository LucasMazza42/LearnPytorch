```python
import yfinance as yf

# Specify the stock symbols and date range
stock_symbols = ['WMT', 'V', 'MSFT', 'JPM', 'XOM']
start_date = '2014-01-01'
end_date = '2024-01-01'

# Dictionary to store each stock's data
stock_data = {}

# Loop through each symbol to download data and store in the dictionary
for symbol in stock_symbols:
    data = yf.download(symbol, start=start_date, end=end_date)
    stock_data[symbol] = data  # Store DataFrame in the dictionary with the symbol as key

# Display the first few rows of each stock's data
for symbol, data in stock_data.items():
    print(f"Data for {symbol}:")
    print(data.head(), "\n")
```

```python
import matplotlib.pyplot as plt

# Plot each stock's Adjusted Close price
for symbol, data in stock_data.items():
    plt.figure(figsize=(10, 5))
    plt.plot(data.index, data['Adj Close'], label=symbol)
    plt.title(f"{symbol} Stock Price Over Time")
    plt.xlabel("Date")
    plt.ylabel("Adjusted Close Price (USD)")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Loop through each stock and plot it against the market
for symbol in stock_symbols:
    # Fetch stock data
    stock_data = yf.download(symbol, start=start_date, end=end_date)['Adj Close']
    
    # Normalize both stock and market data to start at 100
    stock_normalized = stock_data / stock_data.iloc[0] * 100
    market_normalized = market_data / market_data.iloc[0] * 100
    
    # Plot the data
    plt.figure(figsize=(10, 6))
    plt.plot(stock_normalized, label=f'{symbol} (Stock)')
    plt.plot(market_normalized, label='S&P 500 (Market Index)', linestyle='--')
    plt.title(f"{symbol} vs S&P 500")
    plt.xlabel("Date")
    plt.ylabel("Normalized Price (Starting at 100)")
    plt.legend()
    plt.grid(True)
    plt.show()
```
