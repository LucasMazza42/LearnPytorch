```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Stock symbols and dictionary to hold data
stock_symbols = ['WMT', 'V', 'MSFT', 'JPM', 'XOM']
stock_data = {}

# Fetch historical data for each stock
for symbol in stock_symbols:
    stock_data[symbol] = yf.download(symbol, start='2015-01-01', end='2023-01-01')

# Loop through each stock and plot its data
for symbol, data in stock_data.items():
    # Calculate SMAs with different window sizes
    data['SMA_10'] = data['Close'].rolling(window=10).mean()
    data['SMA_50'] = data['Close'].rolling(window=50).mean()
    data['SMA_200'] = data['Close'].rolling(window=200).mean()

    # Identify Golden Crosses and Death Crosses
    golden_crosses = (data['SMA_50'] > data['SMA_200']) & (data['SMA_50'].shift(1) <= data['SMA_200'].shift(1))
    death_crosses = (data['SMA_50'] < data['SMA_200']) & (data['SMA_50'].shift(1) >= data['SMA_200'].shift(1))

    # Plotting
    plt.figure(figsize=(12, 6))
    plt.plot(data['Close'], label='Closing Price', alpha=0.8)
    plt.plot(data['SMA_50'], label='50-Day SMA', linestyle='--', color='green')
    plt.plot(data['SMA_200'], label='200-Day SMA', linestyle='--', color='red')

    # Add markers for Golden Crosses
    plt.scatter(data.index[golden_crosses], data['SMA_50'][golden_crosses], label='Golden Cross', color='gold', marker='^', s=100)

    # Add markers for Death Crosses
    plt.scatter(data.index[death_crosses], data['SMA_50'][death_crosses], label='Death Cross', color='black', marker='v', s=100)

    # Add labels and legend
    plt.title(f'{symbol} Price with Golden Crosses and Death Crosses')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import yfinance as yf

# Define stock symbols and parameters
stock_symbols = ['WMT', 'V', 'MSFT', 'JPM', 'XOM']
look_ahead = 20  # Number of days to look ahead
threshold = 0.05  # 5% price change
false_signals_count = {}  # Dictionary to store false signal counts for each ticker

# Analyze each ticker
for symbol in stock_symbols:
    # Fetch historical data
    data = yf.download(symbol, start='2015-01-01', end='2023-01-01')

    # Calculate SMAs
    data['SMA_50'] = data['Close'].rolling(window=50).mean()
    data['SMA_200'] = data['Close'].rolling(window=200).mean()

    # Identify Golden and Death Crosses
    data['Golden_Cross'] = (data['SMA_50'] > data['SMA_200']) & (data['SMA_50'].shift(1) <= data['SMA_200'].shift(1))
    data['Death_Cross'] = (data['SMA_50'] < data['SMA_200']) & (data['SMA_50'].shift(1) >= data['SMA_200'].shift(1))

    # Initialize counters for wrong signals
    wrong_golden_crosses = 0
    wrong_death_crosses = 0

    # Check Golden Cross signals
    for i in data.index[data['Golden_Cross']]:
        if i + pd.Timedelta(days=look_ahead) in data.index:  # Ensure future date exists
            pct_change = (data.loc[i + pd.Timedelta(days=look_ahead), 'Close'] - data.loc[i, 'Close']) / data.loc[i, 'Close']
            if pct_change < -threshold:  # If price dropped by more than the threshold
                wrong_golden_crosses += 1

    # Check Death Cross signals
    for i in data.index[data['Death_Cross']]:
        if i + pd.Timedelta(days=look_ahead) in data.index:  # Ensure future date exists
            pct_change = (data.loc[i + pd.Timedelta(days=look_ahead), 'Close'] - data.loc[i, 'Close']) / data.loc[i, 'Close']
            if pct_change > threshold:  # If price rose by more than the threshold
                wrong_death_crosses += 1

    # Store total false signals for this ticker
    total_false_signals = wrong_golden_crosses + wrong_death_crosses
    false_signals_count[symbol] = total_false_signals

# Find the ticker with the least false signals
least_false_signals_ticker = min(false_signals_count, key=false_signals_count.get)

# Display the results
print(f"Ticker with the least false signals: {least_false_signals_ticker}")
print("False Signal Counts by Ticker:")
for ticker, count in false_signals_count.items():
    print(f"{ticker}: {count} false signals")

```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch historical data
symbol = 'WMT'
data = yf.download(symbol, start='2020-01-01', end='2023-01-01')

# Calculate Ichimoku components
data['Tenkan_sen'] = (data['High'].rolling(window=9).max() + data['Low'].rolling(window=9).min()) / 2
data['Kijun_sen'] = (data['High'].rolling(window=26).max() + data['Low'].rolling(window=26).min()) / 2
data['Senkou_Span_A'] = ((data['Tenkan_sen'] + data['Kijun_sen']) / 2).shift(26)
data['Senkou_Span_B'] = ((data['High'].rolling(window=52).max() + data['Low'].rolling(window=52).min()) / 2).shift(26)
data['Chikou_Span'] = data['Close'].shift(-26)

# Plot the Ichimoku Cloud
plt.figure(figsize=(12, 8))
plt.plot(data.index, data['Close'], label='Closing Price', color='blue')
plt.plot(data.index, data['Tenkan_sen'], label='Tenkan-sen (9)', color='orange', linestyle='--')
plt.plot(data.index, data['Kijun_sen'], label='Kijun-sen (26)', color='green', linestyle='--')
plt.fill_between(data.index, data['Senkou_Span_A'], data['Senkou_Span_B'],
                 where=data['Senkou_Span_A'] >= data['Senkou_Span_B'], color='lightgreen', alpha=0.5, label='Bullish Cloud')
plt.fill_between(data.index, data['Senkou_Span_A'], data['Senkou_Span_B'],
                 where=data['Senkou_Span_A'] < data['Senkou_Span_B'], color='lightcoral', alpha=0.5, label='Bearish Cloud')
plt.plot(data.index, data['Chikou_Span'], label='Chikou Span (Lagging)', color='purple', linestyle='--')

# Add labels and legend
plt.title(f'{symbol} Ichimoku Cloud')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

def visualize_multi_layer_rnn(hidden_size_1, hidden_size_2, output_size):
    G = nx.DiGraph()
    pos = {}  # Position of each node in the graph
    node_labels = {}  # Labels for nodes

    # Input layer
    G.add_node("x_t", pos=(0, 2))
    node_labels["x_t"] = "x_t"

    # First hidden layer
    for i in range(hidden_size_1):
        node_name = f"h1_{i+1}"
        G.add_node(node_name, pos=(1, 2 - (i * 0.3)))
        node_labels[node_name] = node_name
        G.add_edge("x_t", node_name)  # Connect input to hidden layer 1

    # Second hidden layer
    for j in range(hidden_size_2):
        node_name = f"h2_{j+1}"
        G.add_node(node_name, pos=(2, 2 - (j * 0.5)))
        node_labels[node_name] = node_name
        for i in range(hidden_size_1):
            G.add_edge(f"h1_{i+1}", node_name)  # Connect hidden layer 1 to layer 2

    # Output layer
    for k in range(output_size):
        node_name = f"y_{k+1}"
        G.add_node(node_name, pos=(3, 2))
        node_labels[node_name] = node_name
        for j in range(hidden_size_2):
            G.add_edge(f"h2_{j+1}", node_name)  # Connect hidden layer 2 to output

    # Plot the graph
    pos = nx.get_node_attributes(G, 'pos')
    plt.figure(figsize=(10, 6))
    nx.draw(G, pos, with_labels=False, node_size=800, node_color="lightblue", edge_color="gray")
    nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=10)
    plt.title("Multi-Layer RNN Visualization")
    plt.show()

# Visualize the RNN with the given parameters
visualize_multi_layer_rnn(hidden_size_1=5, hidden_size_2=3, output_size=1)
```

```python
import numpy as np
start_date = "2014-01-01"
end_date = "2024-11-17"
data = yf.download('WMT',start=start_date, end=end_date)

print(data)
```

```python
# Use 'Close' as the target variable
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
close_prices = data['Close'].values

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_close = scaler.fit_transform(close_prices.reshape(-1, 1))

# Create sequences
def create_sequences(data, seq_length):
    x, y = [], []
    for i in range(len(data) - seq_length):
        x.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(x), np.array(y)

SEQ_LENGTH = 14
X, y = create_sequences(scaled_close, SEQ_LENGTH)

# Split into training and testing
split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
print(X_test)
```

```python
# Define the LSTM model

model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(SEQ_LENGTH, 1)),
    LSTM(50),
    Dense(1)  # Predict one value: the next 'Close' price
])

model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()

# Train the model
history = model.fit(X_train, y_train, epochs=20, batch_size=1, validation_data=(X_test, y_test))
```

```python
# Predict on the test set
y_pred = model.predict(X_test)

# Reverse the scaling
y_pred_rescaled = scaler.inverse_transform(y_pred)
y_test_rescaled = scaler.inverse_transform(y_test)

# Compare predictions with actual values
for i in range(len(y_test_rescaled)):
    print(f"Actual: {y_test_rescaled[i][0]:.2f}, Predicted: {y_pred_rescaled[i][0]:.2f}")
```

```python
import matplotlib.pyplot as plt

# Ensure y_test_rescaled and y_pred_rescaled are 1D arrays for plotting
y_test_rescaled_flat = [x[0] for x in y_test_rescaled]
y_pred_rescaled_flat = [x[0] for x in y_pred_rescaled]

# Plot actual vs predicted
plt.figure(figsize=(12, 6))
plt.plot(y_test_rescaled_flat, label="Actual Values", color="blue", linewidth=2)
plt.plot(y_pred_rescaled_flat, label="Predicted Values", color="orange", linestyle="--")
plt.title("LSTM Predictions vs Actual Values")
plt.xlabel("Time Step")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Calculate MSE, RMSE, and MAE
mse = mean_squared_error(y_test_rescaled, y_pred_rescaled)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test_rescaled, y_pred_rescaled)

print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
```
